```python
from dask.distributed import Client

client = await Client(processes=False, asynchronous=True)
```

```python
import xarray as xr
import fsspec

# open dataset as zarr object using fsspec reference file system and xarray
fs = fsspec.filesystem("filecache", 
                       target_protocol='reference', 
                       target_options={
                           'fo': f'./hrrr_hourly/hrrr_kerchunk_cfgrib.json', 
                           'remote_protocol': 's3', 
                           'remote_options': {'anon':True, 'use_ssl': False}
                       })
m = fs.get_mapper("")
ds = xr.open_dataset(m, engine="zarr", backend_kwargs=dict(consolidated=False), chunks={'time': 1})
ds
```

```python
import pyproj
to_xy = pyproj.Transformer.from_crs('epsg:4326', ds.prate.crs, always_xy=True).transform
```

```python
import numpy as np

ROTCON_P = 0.622515
LON_XX_P = -97.5
LAT_TAN_P = 38.5

def wind_speed_dir(u, v, lng):
    """Calculate wind speed (in mph) and direction from u and v components.
    """
    angle2 = ROTCON_P*(lng-LON_XX_P)*0.017453
    sinx2 = np.sin(angle2)
    cosx2 = np.cos(angle2)

    un = cosx2 * u + sinx2 * v
    vn = -sinx2 * u + cosx2 * v

    wind_speed = np.sqrt(un**2 + vn**2) * 2.23694
    wind_dir = (270 - np.arctan2(vn, un) * 180 / np.pi) % 360

    return wind_speed, wind_dir
```

```python
def get_forecast(lat, lng):
    x_sel, y_sel = to_xy(lng, lat)
    
    selected_data = ds.interp(x=x_sel, y=y_sel, method='linear')
    time = selected_data.time.values
    temp = selected_data.tmp.values
    u = selected_data.ugrd.values
    v = selected_data.vgrd.values
    wind_speed, wind_dir = wind_speed_dir(u, v, lng)
    apcp = selected_data.apcp.values

    return {
        'time': time,
        'temp': temp,
        'wind_speed': wind_speed,
        'wind_dir': wind_dir,
        'apcp': apcp
    }
```

```python
forecast = await client.submit(get_forecast, 41.42717976016072, -71.462122760827)
forecast
```
